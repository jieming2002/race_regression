```python
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
path_raw = '../data/raw/'
path_new = '../data/new/'
```

```python
raw_train = pd.read_csv(path_raw + 'tap_fun_train.csv')
raw_train.head()
```

```python
col = 'sr_rss_help_bonus_level'
print('\n%s 属性的不同取值和出现的次数' % col)
print(raw_train[col].value_counts())
```

```python
# sr_rss_help_bonus_level > 0 的是否付费？
tmp_df = raw_train.loc[raw_train['sr_rss_help_bonus_level'] > 0, ['sr_rss_help_bonus_level','pay_price','pay_count','prediction_pay_price']]
tmp_df.head()
```
